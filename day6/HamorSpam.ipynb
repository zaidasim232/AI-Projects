### Data imbalanced

```python
# Generate and plot a synthetic imbalanced classification dataset
from collections import Counter
from sklearn.datasets import make_classification
from matplotlib import pyplot
from numpy import where
# define dataset
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0,
	n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=1)
# summarize class distribution
counter = Counter(y)
print(counter)
# scatter plot of examples by class label
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([('CountVect', CountVectorizer(ngram_range=(1,2))), ('Classifier', RandomForestClassifier())])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['Text'], df['Class'], test_size=0.2, random_state=42)
```

```python
clf.fit(x_train, y_train)
```

```python
tahmin = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, tahmin)
```

```python
print(classification_report(y_test, tahmin))
```

```python
confusion_matrix(y_test, tahmin)
```

```python
from sklearn.svm import SVC
```

```python
svm = Pipeline([('CountVect', CountVectorizer(ngram_range=(1,2))), ('Classifier', SVC())])
```

```python
svm.fit(x_train, y_train)
```

```python
tahmin_svm = svm.predict(x_test)
```

```python
accuracy_score(y_test, tahmin_svm)
```

```python
print(classification_report(y_test, tahmin_svm))
```

```python
confusion_matrix(y_test, tahmin_svm)
```

```python
sns.heatmap(confusion_matrix(y_test, tahmin), annot=True, fmt='d', cmap='Blues')
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
def wc(data, color):
    plt.figure(figsize=(10,10))
    mask = np.array(Image.open('cloud.png'))
    wc = WordCloud(background_color=color, stopwords=STOPWORDS, mask = mask)
    wc.generate(' '.join(data))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
sa= pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)

p = sa("bu telefon modelleri çok kaliteli , her parçası çok özel bence")
print(p)

print(p[0]['label'] == 'LABEL_1')

p = sa("Film çok kötü ve çok sahteydi")
print(p)

print(p[0]['label'] == 'LABEL_1')
```
