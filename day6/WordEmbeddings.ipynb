```python
import gensim
from gensim import corpora
from gensim.models import LdaModel

# Sample documents
documents = [
    "Tennis is my favorite sport to play.",
    "Football is a popular competition in certain country.",
    "There are many athletes currently training for the olympic."
]

# Preprocess documents
texts = [[word for word in document.lower().split()] for document in documents]

dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]


#The LDA model
lda_model = LdaModel(corpus, num_topics=2, id2word=dictionary, passes=15)

topics = lda_model.print_topics()
for topic in topics:
    print(topic)
```

```python
# Imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
import gensim
from gensim import corpora
from gensim.models.ldamulticore import LdaMulticore
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import text
import time # Import the time module
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

# Scrape function
def url_to_transcript(url):
    try:
        page = requests.get(url)
        page.raise_for_status() # Raise an exception for HTTP errors
        soup = BeautifulSoup(page.text, "lxml")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return ""

    text_data = ""

    # Attempt 1: Specific Elementor content div
    content_div = soup.find('div', class_='elementor-widget-theme-post-content')
    if content_div:
        text_data = content_div.get_text(separator=' ', strip=True)
        if text_data:
            print(f'Successfully scraped (elementor-widget-theme-post-content): {url}')
            return text_data
        else:
            print(f'Found elementor-widget-theme-post-content div but no text content for: {url}')

    # Attempt 2: More general content classes/tags as fallbacks
    fallback_classes = ['entry-content clear', 'entry-content', 'post-content']
    for class_name in fallback_classes:
        if not text_data:
            content_div = soup.find('div', class_=class_name)
            if content_div:
                text_data = content_div.get_text(separator=' ', strip=True) # Get all text
                if text_data:
                    print(f'Successfully scraped ({class_name}): {url}')
                    return text_data
                else:
                    print(f'Found {class_name} div but no text content for: {url}')

    # Attempt 3: article tag (general content container)
    if not text_data:
        article_tag = soup.find('article')
        if article_tag:
            text_data = article_tag.get_text(separator=' ', strip=True)
            if text_data:
                print(f'Successfully scraped (article tag): {url}')
                return text_data
            else:
                print(f'Found article tag but no text content for: {url}')

    # Final Fallback: Extract all paragraphs from the entire page
    if not text_data:
        all_paragraphs = soup.find_all('p')
        if all_paragraphs:
            text_data = ' '.join(p.get_text(strip=True) for p in all_paragraphs)
            if text_data:
                print(f'Successfully scraped (all paragraphs fallback): {url}')
                return text_data
            else:
                print(f'Found paragraphs on page but no text content for: {url}')


    print(f'Could not find any suitable content for: {url}')
    return ""

# Custom stopwords
add_stop_words = [
    'like', 'im', 'know', 'just', 'dont', 'thats', 'right', 'people', 'youre', 'got',
    'gonna', 'time', 'think', 'yeah', 'said', 'want', 'say', 'thing', 'going', 'shit',
    'fucking', 'didnt', 'hes', 'shes', 'theyre', 'weve', 'ill', 'youve', 'ive', 'couldnt',
    'wouldnt', 'cant', 'wont', 'one', 'get', 'really', 'see', '♪', 'g', 'okay'
]

# Convert frozenset to list for CountVectorizer
stop_words = list(text.ENGLISH_STOP_WORDS.union(add_stop_words))


# Lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to extract only nouns and adjectives
def nouns_adj(text):
    text = text.lower()
    text = re.sub(r"’", "'", text)
    text = re.sub(r"[^a-zA-Z']", " " , text)
    text = re.sub(r'\s+', ' ', text)
    tokenized = word_tokenize(text)
    tagged = pos_tag(tokenized)
    is_noun_adj = lambda pos: pos.startswith('NN') or pos.startswith('JJ')
    words = [lemmatizer.lemmatize(word) for word, pos in tagged if is_noun_adj(pos) and word not in stop_words]
    return ' '.join(words)

# URLs and comedian names
urls = [
    'http://scrapsfromtheloft.com/2017/05/06/louis-ck-oh-my-god-full-transcript/',
    'http://scrapsfromtheloft.com/2017/04/11/dave-chappelle-age-spin-2017-full-transcript/',
    'http://scrapsfromtheloft.com/2018/03/15/ricky-gervais-humanity-transcript/',
    'https://scrapsfromtheloft.com/comedy/bo-burnham-what-transcript/',
    'http://scrapsfromtheloft.com/2017/05/24/bill-burr-im-sorry-feel-way-2014-full-transcript/',
    'http://scrapsfromtheloft.com/2017/04/21/jim-jefferies-bare-2014-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/02/john-mulaney-comeback-kid-2015-full-transcript/',
    'https://scrapsfromtheloft.com/comedy/hasan-minhaj-homecoming-king-transcript/',
    'http://scrapsfromtheloft.com/2017/09/19/ali-wong-baby-cobra-2016-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/03/anthony-jeselnik-thoughts-prayers-2015-full-transcript/',
    'http://scrapsfromtheloft.com/2018/03/03/mike-birbiglia-my-girlfriends-boyfriend-2013-full-transcript/',
    'http://scrapsfromtheloft.com/2017/08/19/joe-rogan-triggered-2016-full-transcript/'
]

comedians = ['Louis CK', 'Dave Chappelle', 'Ricky Gervais', 'Bo Burnham', 'Bill Burr',
             'Jim Jefferies', 'John Mulaney', 'Hasan Minhaj', 'Ali Wong',
             'Anthony Jeselnik', 'Mike Birbiglia', 'Joe Rogan']

# Scrape and clean
documents = []
for url in urls:
    documents.append(url_to_transcript(url))
    time.sleep(10) # Add a 10-second delay between requests

cleaned_docs = [nouns_adj(doc) for doc in documents if doc]

# Check if cleaned_docs is empty before proceeding with CountVectorizer and LDA
if not cleaned_docs:
    print("No documents were successfully scraped and cleaned. Cannot perform LDA.")
else:
    # Build CountVectorizer
    cv = CountVectorizer(stop_words=stop_words, max_df=0.8)
    data_cv = cv.fit_transform(cleaned_docs)

    # Convert into pandas DataFrame for gensim
    dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names_out())

    # Create corpus for LDA
    corpus = gensim.matutils.Sparse2Corpus(data_cv, documents_columns=False)
    id2word = dict((id, word) for word, id in cv.vocabulary_.items())

    # LDA Model (12 topics for 12 comedians)
    lda_model = LdaMulticore(corpus, num_topics=12, id2word=id2word, passes=50, workers=2, random_state=42)

    # Show topics
    topics = lda_model.print_topics(num_words=10)
    for idx, topic in topics:
        print(f"\n📝 Topic {idx}:\n{topic}")
```
