```python
rating.title.value_counts()
```

```python
matrix_func = rating.pivot_table(index=["user_id"], columns=["title"],values="rating")
```

```python
matrix_func
```

```python
matrix_func["Star Wars (1977)"]
```

```python
Star_W_R = matrix_func["Star Wars (1977)"]
```

```python
Star_W_R.value_counts()
```

```python
matrix_func[["Star Wars (1977)", "101 Dalmatians (1996)"]].corr()
```

```python
sm = matrix_func.corrwith(Star_W_R)
```

```python
sm.sort_values(ascending=False)
```

```python
import numpy as np
```

```python
Matrix_stats = rating.groupby("title").agg({"rating" : [np.size, np.mean]})
```

```python
Matrix_stats.head()
```

```python
pm = Matrix_stats[Matrix_stats["rating"]["size"] > 100]
```

```python
smdata = pd.DataFrame(sm, columns=["similarity"])
```

```python
pm.columns = pm.columns.get_level_values(0)
df = pm.join(smdata)
```

```python
df.sort_values(["similarity"], ascending= False)
```
