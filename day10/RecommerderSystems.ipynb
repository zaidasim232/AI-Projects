```python
rating.title.value_counts()
```

```python
import numpy as np
```

```python
movie_rating = rating.groupby("title").agg({"rating": [np.size, np.sum, np.mean ]})
```

```python
movie_rating
```

```python
en_cok_rating_sirasi = movie_rating.sort_values(by=[("rating", "mean")], ascending=False)
```

```python
en_cok_rating_sirasi
```

```python
toplam_sum_sirasi = en_cok_rating_sirasi.sort_values(by=[("rating", "sum")], ascending=False)
```

```python
toplam_sum_sirasi
```

```python
toplam_rating_sum = toplam_sum_sirasi["rating"]["sum"].sum()
```

```python
toplam_rating_sum
```

```python
toplam_sum_sirasi["rating"]["sum"]
```

```python
toplam_sum_sirasi["percentage"] = toplam_sum_sirasi["rating"]["sum"].div(toplam_rating_sum) * 100
```

```python
toplam_sum_sirasi["rank"] = toplam_sum_sirasi["percentage"].rank(ascending=False)
```
