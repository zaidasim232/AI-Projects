# Normalizing Data and Principle Component Analysis (PCA)

```python
import pandas as pd

df = pd.read_csv('PastHires.csv')
```

```python
df.head()
```

```python
df['Employed?'] = df['Employed?'].map({'Y': 1, 'N': 0})
```

```python
df
```

```python
df = df.replace({'Y' : 1, 'N' : 0, 'BS': 1, 'MS': 2, 'PhD': 3})
```

```python
df
```
