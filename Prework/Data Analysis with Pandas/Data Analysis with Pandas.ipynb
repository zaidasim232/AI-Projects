
#### Additional Dataframe Operations

16. Perform a `merge` using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as `clients`.
17. Perform a `merge` using `clients` and "purchase_log.csv" and limit the subset to only clients who made purchases. Assign the resulting dataframe as `detailed_sales`.
18. Use `groupby` to find the client who spent the most money on purchases. Determine how much he/she spent. HINT: save the intermediate dataframe from using `groupby` as `spenders` before applying slicing to determine the client who spent the most money on purchases.
19. (BONUS) Modify the answer to problem 3 slightly to determine exactly what items where purchased by the top spending client.

```python
df = pd.read_csv('client_list.csv')
df.columns = [x.upper() for x in df.columns]
df2 = pd.read_csv('customer_id_list.csv')
client = pd.merge(left=df, right=df2, how='left', on=['LAST_NAME','FIRST_NAME','GENDER','AGE'])
client.head()
```

```python
purchase_log = pd.read_csv('purchase_log.csv')
datailed_sale = pd.merge(left=client, right=purchase_log, how='inner', on='CUSTOMER_ID')
datailed_sale.head()
```

```python
spenders = datailed_sale.groupby(['FIRST_NAME','LAST_NAME','CUSTOMER_ID'])['PRICE'].sum().reset_index()
spenders = spenders.sort_values(by='PRICE', ascending=False)
top_spender = spenders.iloc[0]
top_spender

#spenders = datailed_sale.groupby(['CUSTOMER_ID','FIRST_NAME','LAST_NAME','GENDER','AGE'], as_index=False)['PRICE'].sum()
#top_spender = spenders[spenders.PRICE==spenders.PRICE.max()]
#top_spender
```

```python
top_spender_id = top_spender.reset_index(drop=True).loc[0,'CUSTOMER_ID']
top_spender_items = datailed_sale[datailed_sale['CUSTOMER_ID'] == top_spender_id][['ITEM_DESCRIPTION','PRICE']]
top_spender_items
```
