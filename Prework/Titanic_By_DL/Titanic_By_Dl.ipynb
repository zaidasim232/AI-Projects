```python
df = pd.concat([df_train, df_test], ignore_index=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
del df['Cabin']
```

```python
df["Embarked"].value_counts()
```

```python
df.fillna({"Embarked": df["Embarked"].mode()[0]}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Title"] = df["Name"].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df["Title"].value_counts()
```

```python
df["Title"].unique()
```

```python
df["Title"] = df["Title"].replace(['Rev','Dr','Col','Major','Sir','Don','Capt','Jonkheer','Master'], 'Mr')
```

```python
df['Title'] = df['Title'].replace(['Lady','Countess','Dona','Mme','Ms', 'Mlle'], 'Mrs')
```

```python
df["Title"].value_counts()
```

```python
df[df['Title']=='Mr']
```

```python
df.sample(10)
```

```python
#df['Age'].fillna(df[df['Title']=='Mr']['Age'].mean(), inplace=True)

from itertools import groupby


df['Age'] = df['Age'].fillna(df.groupby('Title')['Age'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
```

```python
df.head()
```

```python
del df['Ticket'], df['Title']
```

```python
del df['Name']
```

```python
del df['Embarked']
```

```python
df.head()
```

```python
train = df.iloc[:891]
test = df.iloc[891:]
```

```python
test.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(train.drop(['Survived','PassengerId'], axis=1))
y = train['Survived']
```

```python
test = test.drop(['Survived','PassengerId'], axis=1)
```

```python
x.shape
```

```python
test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential(
    [

        Dense(32, activation='relu', input_shape=(x.shape[1],)),
        Dense(64, activation='relu'),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ]
)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history = model.fit(x, y, epochs=100, batch_size=32, validation_split=0.10, verbose=1, callbacks=[early_stopping])
```

```python
loss, accuracy = model.evaluate(x, y)
```

```python
accuracy
```

```python
tahmin = model.predict(test).astype('int32').flatten()
```

```python
tahmin
```

```python
sonuc = pd.DataFrame({'PassengerId': df_test['PassengerId'], 'Survived': tahmin})
```
